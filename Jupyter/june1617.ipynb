# Matplotlib Library - Data Visualization
# Graph - animated, attractive charts/plots
#### pyplot
#### line chart(time trends)
#### bar chart(count categorial values)
#### histogram(distributions)
#### Pie Chart(area section)
#### scatter plot(relationship b/w two variables)
#### Box Plot (for checking inlayer and outlayer values)
#### Heatmap (correlation)

```python
import matplotlib.pyplot as plt
```

Line Chart

```python
x=[1,2,3,4,5,6,7,8]
y=[12,23,34,45,56,67,78,89]
# line chart - Line
# bar chart - bar
# histogram - hist
# Pie Chart - pie
# scatter plot - scatter plot
```

```python
plt.plot(x,y)   # To plot chart
plt.title("LineChart")  # Chart heading 
plt.xlabel("X-axis")    # X-axis
plt.ylabel("Y-axis")    # Y-axis
plt.show()              # To show graph
```

Common line style - Dotted, solid, dashed, dash-dot
Markers - circle(o), square(s)

Bar Graph

```python
plt.bar(x,y)
plt.title("BarGraph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Pie Chart

```python
plt.pie(x,labels=y)
plt.title("PieChart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Histogram

```python
plt.hist(y)
plt.title("Histogram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Edge Color
```

scatter Plot

```python
plt.scatter(x,y)
plt.title("scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
