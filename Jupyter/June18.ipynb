plots

distribution plot

```python
# histogram
sns.histplot(data=df,x="total_bill",bins=10,kde=True)
plt.show()
```

```python
# KDE(kernel density estimator)
sns.kdeplot(data=df,x="total_bill",fill=True)
plt.show()
```

relational plots

```python
# scatter plot
sns.scatterplot(data=df,x="total_bill",y="tip",hue="sex",palette="Blues")
plt.show()
```

```python
# line plot
sns.lineplot(data=df,x="total_bill",y="tip",hue="sex")
plt.show()
```

categorical plots

```python
# bar plot
sns.barplot(data=df,x="day",y="total_bill")
plt.show()
```

```python
# count plot
sns.countplot(data=df,x="sex")
plt.show()
```

```python
# box plot
sns.boxplot(data=df,x="day",y="total_bill")
plt.show()
```

```python
# violin plot
sns.violinplot(data=df,x="day",y="total_bill")
plt.show()
```

```python
# correlation matrix
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# pair plot
sns.pairplot(data=df,hue="sex")
plt.show()
```

```python
# joint plot
sns.jointplot(data=df,x="total_bill",y="tip",kind="reg")
plt.show()
```
